```python
import tensorflow as tf
import numpy as np
```

```python
EPOCHS = 6*1e5
lr = 0.1
batch_size = 256
weight_decay = 0.0001
num_class = 7
```

```python
class Resnet152(tf.keras.Model):
    def __init__(self):
        super(Resnet152, self).__init__()
    def conv1(self, inputs, training):
        x = tf.keras.layers.Conv2D(64, 7, stride = 2)
        x = tf.keras.layers.BatchNormalization()(x, training = training)
        return x
    def conv2_x(self, inputs, training):
        x = tf.keras.layers.MaxPool2D(3, stride = 2)(inputs)
        inputs = x
        for i in range(3):
            if i == 0:
                x = tf.keras.layers.Conv2D(64, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(64, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(256, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                inputs = tf.keras.layers.Conv2D(256, 1, padding = 'SAME')(inputs)
                inputs = tf.keras.layers.BatchNormalization()(inputs, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
            else:
                x = tf.keras.layers.Conv2D(64, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(64, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(256, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
        return x
    
    def conv3_x(self, inputs, training):
        x = tf.keras.layers.MaxPool2D(3, stride = 2)(inputs)
        inputs = x
        for i in range(4):
            if i == 0:
                x = tf.keras.layers.Conv2D(128, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(128, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(512, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                inputs = tf.keras.layers.Conv2D(512, 1, padding = 'SAME')(inputs)
                inputs = tf.keras.layers.BatchNormalization()(inputs, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
            else:
                x = tf.keras.layers.Conv2D(128, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(128, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(512, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
        return x
    
    def conv4_x(self, inputs, training):
        x = tf.keras.layers.MaxPool2D(3, stride = 2)(inputs)
        inputs = x
        for i in range(23):
            if i == 0:
                x = tf.keras.layers.Conv2D(256, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(256, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(1024, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                inputs = tf.keras.layers.Conv2D(1024, 1, padding = 'SAME')(inputs)
                inputs = tf.keras.layers.BatchNormalization()(inputs, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
            else:
                x = tf.keras.layers.Conv2D(256, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(256, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(1024, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
        return x
    
    def conv5_x(self, inputs, training):
        x = tf.keras.layers.MaxPool2D(3, stride = 2)(inputs)
        inputs = x
        for i in range(3):
            if i == 0:
                x = tf.keras.layers.Conv2D(512, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(512, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(2048, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                inputs = tf.keras.layers.Conv2D(2048, 1, padding = 'SAME')(inputs)
                inputs = tf.keras.layers.BatchNormalization()(inputs, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
            else:
                x = tf.keras.layers.Conv2D(512, 1, padding = 'SAME')(inputs)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(512, 3, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = tf.keras.layers.Activation('relu', x)
                x = tf.keras.layers.Conv2D(2048, 1, padding = 'SAME')(x)
                x = tf.keras.layers.BatchNormalization()(x, training = training)
                x = x+inputs
                x = tf.keras.layers.Activation('relu', x)
                inputs = x
        return x
    
    def fully_connected(self, inputs):
        x = tf.nn.avg_pool(inputs)
        x = tf.keras.layers.Dense(num_class, 'softmax')(x)
        return x
```
